### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 10
x = np.linspace(0, 10, 100)

# Create y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y)

# Add a title and labels for the axes
plt.title('Square of x (x^2)')
plt.xlabel('x values')
plt.ylabel('y values (x^2)')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```
